```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = 50
T = 1/f
omega = 2*np.pi*f
U = 230
t = np.arange(0, 2*T, T/20)
# alternativ:
t = np.linspace(0, 2*T, 40)
t
```

```python
u = 230*np.sqrt(2)*np.sin(omega*t)
```

```python
# Demonstrasjonsdata
x1 = np.linspace(0.0, 5.0)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
x2 = np.linspace(0.0, 2.0)
y2 = np.cos(2 * np.pi * x2)
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1)
fig.suptitle('subplot demo')

ax1.plot(x1, y1, 'o-')
ax1.set_ylabel('Dempa oscillering')

ax2.plot(x2, y2, '.-')
ax2.set_xlabel('tid (s)')
ax2.set_ylabel('Udempa oscillering')

plt.show()
```
